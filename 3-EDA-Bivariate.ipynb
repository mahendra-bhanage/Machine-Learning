# 8. Lineplot (Numerical-Numerical)

#### Q: Yearwise how many passengers have travelled?

```python
flights.head()
```

```python
new = flights.groupby('year').sum().reset_index()
new
```

```python
sns.lineplot(new['year'],new['passengers'])
```

```python
h = flights.pivot_table(values='passengers',index='month',columns='year')
h
```

```python
sns.heatmap(h)
```

- heatmap showa as year passes the number of passengers increased
- no. of passenges count is more in the summer months(may to sep)
